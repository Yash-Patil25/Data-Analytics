```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Cars93.csv")
```

```python
avg_price = df.groupby("Manufacturer")["Price"].mean()

plt.figure(figsize=(12,6))
avg_price.plot(kind="bar")

plt.xlabel("Manufacturer")
plt.ylabel("Average Price ($)")
plt.title("Average Car Price by Manufacturer")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

for t in df["Type"].unique():
    subset = df[df["Type"] == t]
    plt.plot(subset.index, subset["Horsepower"], label=t)

plt.title("Horsepower Trend by Car Type")
plt.xlabel("Index")
plt.ylabel("Horsepower")
plt.legend()
plt.grid()
plt.show()
```

```python
type_counts = df["Type"].value_counts()

plt.figure(figsize=(8,8))
plt.pie(type_counts, labels=type_counts.index, autopct="%1.1f%%")
plt.title("Proportion of Car Types")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["Horsepower"], df["MPG.city"])

plt.xlabel("Horsepower")
plt.ylabel("MPG (City)")
plt.title("Horsepower vs MPG (City)")
plt.grid()
plt.show()
```

```python
df = pd.read_csv("tips.csv")
```

```python
plt.figure(figsize=(8,6))

for gender in df["sex"].unique():
    subset = df[df["sex"] == gender]
    plt.scatter(subset["total_bill"], subset["tip"], label=gender)

plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter Plot of Tip vs Total Bill by Sex")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df["total_bill"], bins=50)

plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.title("Histogram of Total Bill (50 Bins)")
plt.grid()
plt.show()
```

```python
import numpy as np

corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,8))
plt.imshow(corr, cmap="coolwarm", interpolation="nearest")
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap of Tips Dataset")
plt.tight_layout()
plt.show()
```
