```python
#Q8 Create two arrays of shape (2, 3): arr1 and arr2. Perform element-wise addition,subtraction, multiplication, and division on them.
import numpy as np

arr1 = np.array([[1, 2, 3],
                 [4, 5, 6]])

arr2 = np.array([[6, 5, 4],
                 [3, 2, 1]])

print("arr1:\n", arr1)
print("arr2:\n", arr2)

print("\nAddition:\n", arr1 + arr2)
print("\nSubtraction:\n", arr1 - arr2)
print("\nMultiplication:\n", arr1 * arr2)
print("\nDivision:\n", arr1 / arr2)
```

```python
#Q9 Use NumPy's built-in functions to compute the sum, mean, and standard deviation of a given array.
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

total_sum = np.sum(arr)
mean_value = np.mean(arr)
std_deviation = np.std(arr)

print("Array:", arr)
print("Sum:", total_sum)
print("Mean:", mean_value)
print("Standard Deviation:", std_deviation)
```

```python
#Q10 Create a 2x2 matrix A and a 2x2 matrix B. Perform matrix multiplication using np.dot() or the @ operator.
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)

result_dot = np.dot(A, B)

result_at = A @ B

print("\nMatrix Multiplication using np.dot():\n", result_dot)
print("\nMatrix Multiplication using @ operator:\n", result_at)
```

```python
#Q11 Find the transpose of matrix A and matrix B.
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

A_T = A.T
B_T = B.T

print("Matrix A:\n", A)
print("Transpose of A:\n", A_T)

print("\nMatrix B:\n", B)
print("Transpose of B:\n", B_T)

```
