```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("tips.csv")

df.head()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Boxplot of Total Bill by Day")
plt.show()
```

```python
sns.pairplot(df, hue="time")
plt.show()
```

```python
df["sex_label"] = df["sex"].map({0: "Male", 1: "Female"})

plt.figure(figsize=(7,5))
sns.scatterplot(x="total_bill", y="tip", hue="sex_label", data=df)
plt.title("Scatter Plot of Tip vs Total Bill")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["total_bill"], bins=50)
plt.title("Histogram of Total Bill (50 Bins)")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()

```

```python
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(8,5))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
