```python
#Q3 Perform the following operations on an array of mobile phones prices 6999, 7500, 11999, 27899, 14999, 9999.
```

```python
import numpy as np
```

```python
# a. Create a 1D-array of mobile phone prices
prices = np.array([6999,7500,11999,27899,14999,9999])
print("Oringinal Prices:",prices)
```

```python
# b. Convert this array to float type
prices = prices.astype(float)
print("Float Prices:",prices)
```

```python
# c. Append a new mobile having price of 13999
prices = np.append(prices, 13999)
print("After Appending 13999:", prices)
```

```python
# d. Reverse this array
prices = prices[::-1]
print("Reversed Prices:", prices)
```

```python
# e. Apply GST of 18% and update
prices = prices * 1.18
print("After Applying 18% GST:", prices)
```

```python
# f. Sort array in descending order
prices_desc = np.sort(prices)[::-1]
print("Sorted (Descending):", prices_desc)
```

```python
# g. Average mobile phone price
average_price = np.mean(prices)
print("Average Price:", average_price)
```

```python
# h. Difference between max and min price
difference = np.max(prices) - np.min(prices)
print("Difference (Max - Min):", difference)
```
